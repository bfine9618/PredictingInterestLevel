```python
import numpy as np
import pandas as pd

import requests as re
```

```python
apt = ['apartment', 'apt']
condo = ['condominium', 'condo']
walkUp = ['walk_up', 'walk-up', 'walkup', 'walk up']
studio = ['studio']
ph = ['ph', 'penhouse']
townhome = ['townhome', 'duplex', 'townhouse']
loft = ['loft']

types = [apt, condo, walkUp, studio, ph, townhome, loft]
```

```python
df = pd.read_json('./rawData/cleaned.json')
```

```python
def unitType(x):
    homeType = {    
        }
    for lst in types:
        homeType[lst[0]] = False
    
    for lst in types:
        for w in lst:
            if w in x:
                homeType[lst[0]] = True
                return homeType
    return None
```

```python
df['type'] = df['description'].apply(lambda x : unitType(x))
```

```python
df.info()
```

```python
df = pd.concat([df.drop(['type'], axis=1), df['type'].apply(pd.Series)], axis=1)
```

```python
typed = df[~pd.isnull(df['apartment'])]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(typed.drop('interestVal', axis=1), 
                                                    typed['interestVal'], test_size=0.33, random_state=42)
```

```python
train_typed = pd.concat([X_train, y_train], axis=1)
```

```python
train_typed.to_json('./typedData/train_typed.json')
X_test.to_json('./typedData/testX.json')
y_test.to_json('./typedData/y_test.json')
```
